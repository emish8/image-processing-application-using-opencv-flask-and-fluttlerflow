```python

#Aneri- Morphology 


# Read function
#def readImage(img):
 #   return cv2.imread(img)

#img = cv2.imread("img.jpg")

# Kernel Function
def kernelFunc(size):
    return np.ones((size, size), np.uint8)


# Orignal Image
def orignalImage(img):
    rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return rgb_image


# Erosion Matrix Function
def matrixEro(img, size_of_kernel=3, no_of_iterations=5):
    orignal_image = orignalImage(img)
    erosion = cv2.erode(orignal_image, kernel=kernelFunc(size_of_kernel), iterations=no_of_iterations)
    return erosion


# Dilation Matrix Function
def matrixDil(img, size_of_kernel=3, no_of_iterations=5):
    orignal_image = orignalImage(img)
    dilation = cv2.dilate(orignal_image, kernel=kernelFunc(size_of_kernel), iterations=no_of_iterations)
    return dilation


# Gradient Matrix Function
def matrixGrad(img, size_of_kernel=3, no_of_iterations=5):
    gradient = matrixEro(img, size_of_kernel, no_of_iterations) - matrixDil(img, size_of_kernel, no_of_iterations)
    return gradient


# Opening  Matrix Function
def matrixOpening(img, size_of_kernel=3, no_of_iterations=5):
    erosion_image = matrixEro(img, size_of_kernel, no_of_iterations)
    opening_image = cv2.dilate(erosion_image, kernel=kernelFunc(size_of_kernel), iterations=no_of_iterations)
    return opening_image


# Closing Matrix Function
def matrixClosing(img, size_of_kernel=3, no_of_iterations=5):
    orignal_image = orignalImage(img)
    dilation_image = matrixDil(img, size_of_kernel, no_of_iterations)
    closing_image = cv2.erode(dilation_image, kernel=kernelFunc(size_of_kernel), iterations=no_of_iterations)
    return closing_image


# Black Hat Matrix Function
def matrixBlackHat(img, size_of_kernel=3, no_of_iterations=5):
    orignal_image = orignalImage(img)
    closing_img = matrixClosing(img, size_of_kernel, no_of_iterations)
    black_hat = closing_img - orignal_image
    return black_hat


# Top Hat Matrix Function
def matrixTopHat(img, no_of_iterations=5, size_of_kernel=3, no_tophat_iterations=6):
    orignal_image = orignalImage(img)
    opening_image = matrixOpening(img, no_of_iterations, size_of_kernel)
    for i in range(0, no_tophat_iterations):
        top_hat = orignal_image - opening_image
        top_hat
    return top_hat


# Plot All in one Function
def plotAll(img, size_of_kernel=3, no_of_iterations=5, no_tophat_iterations=6):
    orig_img = orignalImage(img)
    ero_img_1 = matrixEro(img, size_of_kernel, no_of_iterations)
    dil_img_2 = matrixDil(img, size_of_kernel, no_of_iterations)
    gra_img_3 = matrixGrad(img, size_of_kernel, no_of_iterations)
    ope_img_4 = matrixOpening(img, size_of_kernel, no_of_iterations)
    clo_img_5 = matrixClosing(img, size_of_kernel, no_of_iterations)
    bla_img_6 = matrixBlackHat(img, size_of_kernel, no_of_iterations)
    top_img_7 = matrixTopHat(img, no_of_iterations, size_of_kernel, no_tophat_iterations)
    imglist = [orig_img, ero_img_1, dil_img_2, gra_img_3, ope_img_4, clo_img_5, bla_img_6, top_img_7]

    fig, ax = plt.subplots(4, 2, figsize=(8, 8), facecolor="w", tight_layout=True)
    plt.subplots_adjust(left=0.01,
                        bottom=0.01,
                        right=0.9,
                        top=0.9,
                        wspace=0.4,
                        hspace=0.4)

    ax[0, 0].imshow(imglist[0])
    ax[0, 0].set_title("Orignal Image")
    ax[0, 0].set_xticklabels(labels="", rotation=90, c="w", fontweight="bold", fontsize=12)
    ax[0, 0].set_yticklabels(labels="", rotation=360, c="w", fontweight="bold", fontsize=12)

    ax[0, 1].imshow(imglist[1])
    ax[0, 1].set_title("Erosion Image")
    ax[0, 1].set_xticklabels(labels="", rotation=90, c="w", fontweight="bold", fontsize=12)
    ax[0, 1].set_yticklabels(labels="", rotation=360, c="w", fontweight="bold", fontsize=12)

    ax[1, 0].imshow(imglist[2])
    ax[1, 0].set_title("Dilation Image")
    ax[1, 0].set_xticklabels(labels="", rotation=90, c="w", fontweight="bold", fontsize=12)
    ax[1, 0].set_yticklabels(labels="", rotation=360, c="w", fontweight="bold", fontsize=12)

    ax[1, 1].imshow(imglist[3])
    ax[1, 1].set_title("Gradient Image")
    ax[1, 1].set_xticklabels(labels="", rotation=90, c="w", fontweight="bold", fontsize=12)
    ax[1, 1].set_yticklabels(labels="", rotation=360, c="w", fontweight="bold", fontsize=12)

    ax[2, 0].imshow(imglist[4])
    ax[2, 0].set_title("Opening Image")
    ax[2, 0].set_xticklabels(labels="", rotation=90, c="w", fontweight="bold", fontsize=12)
    ax[2, 0].set_yticklabels(labels="", rotation=360, c="w", fontweight="bold", fontsize=12)

    ax[2, 1].imshow(imglist[5])
    ax[2, 1].set_xticklabels(labels="", rotation=90, c="w", fontweight="bold", fontsize=12)
    ax[2, 1].set_yticklabels(labels="", rotation=360, c="w", fontweight="bold", fontsize=12)
    ax[2, 1].set_title("Closing Image")

    ax[3, 0].imshow(imglist[6])
    ax[3, 0].set_xticklabels(labels="", rotation=90, c="w", fontweight="bold", fontsize=12)
    ax[3, 0].set_yticklabels(labels="", rotation=360, c="w", fontweight="bold", fontsize=12)
    ax[3, 0].set_title("Black Hat Image")

    ax[3, 1].imshow(imglist[7])
    ax[3, 1].set_xticklabels(labels="", rotation=90, c="w", fontweight="bold", fontsize=12)
    ax[3, 1].set_yticklabels(labels="", rotation=360, c="w", fontweight="bold", fontsize=12)
    ax[3, 1].set_title("Top Hat Image")

    plt.savefig("abc.png", dpi=1000, bbox_inches='tight')
    img_rtr = cv2.imread("abc.png")
    img_rtr = cv2.cvtColor(img_rtr, cv2.COLOR_BGR2RGB)
    os.remove("abc.png")

    return img_rtr




```

```python
#erosion
def erosion(img):
    k = np.ones((5,5)).astype(int)
    erosion = cv2.erode(img, kernel = k, iterations= 1)
    return  erosion
```

```python
#kinjal- bilatral
def bilateral_filter(img):
    #img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    bilateral = cv2.bilateralFilter(img, 15, 85, 80)
    #output=plt.imshow(bilateral)
    return bilateral
```

```python
#vishwa patel- smoothing filter
def smoothing(img,fltr= "MedianBlur"):  

    #img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    if fltr == "GaussianBlur":
        title = "GaussianBlur"
        output = cv2.GaussianBlur(img,(5,5),sigmaX=0)
    else :
        title = "medianBlur"
        output = cv2.medianBlur(img,ksize=5)
    return output
    
```

```python
#Vibhuti- threshold
def thresh(img):
    
    #img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    grayimage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, binary = cv2.threshold(grayimage, 100, 255, cv2.THRESH_BINARY)
    ret, binaryinv = cv2.threshold(grayimage, 100, 255, cv2.THRESH_BINARY_INV)
    ret, trunc = cv2.threshold(grayimage, 100, 255, cv2.THRESH_TRUNC)
    ret, tozero = cv2.threshold(grayimage, 100, 255, cv2.THRESH_TOZERO)
    ret, tozeroinv = cv2.threshold(grayimage, 100, 255, cv2.THRESH_TOZERO_INV)
    adaptive_thresh1 = cv2.adaptiveThreshold(grayimage, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
    adaptive_thresh2 = cv2.adaptiveThreshold(grayimage, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

    
    thresh_output= binaryinv
    
    return thresh_output


```

```python
#devangee- harris corner detection
def HarrisCorner(img):
    operatedImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    operatedImage = np.float32(operatedImage)
    dest = cv2.cornerHarris(operatedImage, 2, 5, 0.07)
    dest = cv2.dilate(dest, None)
    img[dest > 0.01 * dest.max()]=[0, 0, 255]
    #output=plt.imshow(dest, cmap='gray')
    harris_output=img
    return harris_output
```

```python
# MUSKAN- Canny Edge Detection
def CannyEdge(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    blur = cv2.GaussianBlur(gray, (3, 3), 0) # help of GussianBlur
    edge = cv2.Canny(blur, 10, 100)
    #output=plt.imshow(edge)
    return edge
```

```python
#Vishwa Soni : Image enhancement Contrast Streching 

def preprocessing(img):
    # resizing using aspect ratio intact and finding the circle
    # reduce size retain aspect ratio intact
    # invert BGR 2 RGB
    RGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    Ig = RGB[:, :, 2]
    [w,h] = np.shape(Ig)
    r=1200.0/Ig.shape[1]
    dim=(1200,int(Ig.shape[0]*r))
    rz = cv2.resize(Ig,dim,interpolation=cv2.INTER_AREA)
    #  convert in to float and get log trasform for contrast streching
    g = 0.2 * (np.log(1 + np.float32(rz)))
    # change into uint8
    cvuint = cv2.convertScaleAbs(g)
    # cvuint8.dtype
    ret, th = cv2.threshold(cvuint, 0, 255, cv2.THRESH_OTSU)
    ret1,th1 = cv2.threshold(Ig,0,255,cv2.THRESH_OTSU)
    # closeing operation
    # from skimage.morphology import disk
    # from skimage.morphology import erosion, dilation, opening, closing, white_tophat
    # selem = disk(30)
    # cls = opening(th, selem)
    # plot_comparison(orig_phantom, eroded, 'erosion')
    # in case using opencv
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (35,35))
    cls = cv2.morphologyEx(th, cv2.MORPH_OPEN, kernel)
    Im = cls*rz # the mask with resize image
    # cv2.imwrite('mynew.jpg', mask)
    #output=plt.imshow(g,cmap='gray')
    return Im
```

```python
#Khushi Sagar- DOWNSAMPLE-Lower Resolution

def lower_reso(img):
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    grayimage = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
    lower_reso = cv2.pyrDown(rgb)
    #output=plt.imshow(lower_reso, cmap="gray")
    return lower_reso
```

```python
#FORAM- Rotation

def combine (img):
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    grayimage = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)

    res = cv2.resize(grayimage,None,fx=2, fy=4, interpolation = cv2.INTER_CUBIC)

    # height, width = img.shape[:2]
    # res = cv2.resize(grayimage,(2*width, 4*height), interpolation = cv2.INTER_CUBIC)

    rows, cols = grayimage.shape

    M1 = np.float32([[1,0,-100],[0,1,-100]])
    dst1 = cv2.warpAffine(grayimage, M1, (cols, rows))

    M2 = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
    dst2 = cv2.warpAffine(grayimage,M2,(cols,rows))

    pts11 = np.float32([[50,50],[200,50],[50,200]])
    pts12 = np.float32([[10,100],[200,50],[100,250]])
    M3 = cv2.getAffineTransform(pts11,pts12)
    dst3 = cv2.warpAffine(grayimage,M3,(cols,rows))


    pts21 = np.float32([[50,50],[550,50],[50,550],[550,550]])
    pts22 = np.float32([[0,0],[500,0],[0,500],[500,500]])
    M4 = cv2.getPerspectiveTransform(pts21,pts22)
    dst4 = cv2.warpPerspective(grayimage,M4,(500,500))

    
    #output=plt.imshow(dst2, cmap="gray"),plt.title("Rotation")
    #plt.xticks([]), plt.yticks([])
    return dst2
```

```python
#Aarti -Laplacian sharpen filter
def laplacian(img):
    sharpen_kernel = np.array([[0, -1, 0],
                          [-1, 5, -1],
                          [0, -1, 0]])
    laplacian = cv2.filter2D(src=img,ddepth=-1,kernel = sharpen_kernel)
    #plt.figure(figsize=(5,3))
    #output=plt.imshow(laplacian)
    return laplacian
```

```python
#KASHISH- RGB to GRAY
def RgbGray(img, dist_val = 175, max_val=50):
       
        fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        R, G, B = fix_img[:,:,0], fix_img[:,:,1],fix_img[:,:,2]
    
        grayscale_weighted_img = 0.299 * R + 0.5878 * G + 0.114 * B
        return grayscale_weighted_img
```

```python
#Bansi - (Hough transform)
import math 

def LineDetection(img):
    dst = cv2.Canny(img, 50, 200, None, 3)
    
    # Copy edges to the images that will display the results in BGR
    cdst = cv2.cvtColor(dst, cv2.COLOR_GRAY2BGR)
    cdstP = np.copy(cdst)
    
    lines = cv2.HoughLines(dst, 1, np.pi / 180, 150, None, 0, 0)
    
    if lines is not None:
        for i in range(0, len(lines)):
            rho = lines[i][0][0]
            theta = lines[i][0][1]
            a = math.cos(theta)
            b = math.sin(theta)
            x0 = a * rho
            y0 = b * rho
            pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
            pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))
            cv2.line(cdst, pt1, pt2, (0,0,255), 3, cv2.LINE_AA)
    
    
    linesP = cv2.HoughLinesP(dst, 1, np.pi / 180, 50, None, 50, 10)
    
    if linesP is not None:
        for i in range(0, len(linesP)):
            l = linesP[i][0]
            cv2.line(cdstP, (l[0], l[1]), (l[2], l[3]), (0,0,255), 3, cv2.LINE_AA)
    
    return cdstP
```

```python
#Jinal (Convolution function)
def gaussianBlur(img):
    
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gaussianBlur = cv2.GaussianBlur(img,(7,7),0)
    return gaussianBlur
```

```python
#Zeel Rathi Watershed Function

def watershed(img):
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU) #thresholding the imae using otsu's binarization
    kernel = np.ones((3, 3), np.uint8)
    thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2) # Removing small regions
    extendedFg = cv2.dilate(thresh, kernel, iterations=2)
    distance = cv2.distanceTransform(thresh, cv2.DIST_L2, 5)# foreground centers
    ret, fgCenters = cv2.threshold(distance, 0.7 * distance.max(), 255, 0)
    fgCenters = np.uint8(fgCenters)  # unknown regions subtracting the foreground centers to the extended foreground
    unknown = cv2.subtract(extendedFg, fgCenters)
    ret, markers = cv2.connectedComponents(fgCenters) #defining the markers
    markers = markers + 1
    markers[unknown == 255] = 0 # marking the unknown regions with zero
    markers = cv2.watershed(img, markers)  # applying the method
    img[markers == -1] = [255, 0, 0]  # draw the boundary regions marked with -1 , ont the original image
   
    #output=plt.imshow(markers),plt.title("watersheded image")
   
    return extendedFg
```

```python
app = Flask(__name__)
run_with_ngrok(app)  



def read_img(url):
  req = urllib.request.urlopen(url)
  arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
  img = cv2.imdecode(arr, -1) 

  return  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  

@app.route('/', methods=['POST','GET'] )
def convert():
   
    if(request.method == "POST"):
      print('got post request')
      data = request.get_json()
      print(data)
      try:
        image_path = data['img_path']
        img =  read_img(image_path)
       
      except:
        print('image path error')

      try:
        opr = data['operation']
        print(opr)

      except:
        print('operation error')

      if opr == 'erosion':
        output = erosion(img)
        print("erosion called", output)
        plt.imshow(output)
        output_url = upload(output)

        return  {"output": output_url,"status":True}

        
      #Aneri - Morphology
      elif opr == 'morphology':
        output = plotAll(img, size_of_kernel=3, no_of_iterations=5, no_tophat_iterations=6)
        print("morphology called", output)
        #plt.imshow(output)
        output_url = upload(output)

        return  {"output": output_url,"status":True}


      
      # Kinjal-bilateral
      elif opr == 'bilateral':
        output = bilateral_filter(img)
        print("bilateral called", output)
        #plt.imshow(output)
        output_url = upload(output)

        return  {"output": output_url,"status":True}
      
      # Vishwa Patel - SMOOTHING 
      elif opr == 'smoothing':
        output = smoothing(img)
        print("smoothing called", output)
        output_url = upload(output)

        return  {"output": output_url,"status":True}
        
      #Vibhuti- threshold
      elif opr == 'thresh':
        output = thresh(img)
        print("thresh called", output)
        output_url = upload(output)

        return  {"output": output_url,"status":True}

      #devangee- harris corner detection
      elif opr == 'harris':
        output = HarrisCorner(img)
        print("harris called", output)
        output_url = upload(output)

        return  {"output": output_url,"status":True} 

   
      #MUSKAN- Canny Edge Detection
      elif opr == 'canny':
        output = CannyEdge(img)
        print("canny called", output)
        output_url = upload(output)
        print('output_url',output_url)
        return  {"output": output_url,"status":True} 

      

     
      #Vishwa Soni : Image enhancement Contrast Streching
      elif opr == 'preprocessing':
        output = preprocessing(img)
        print("preprocessing called", output)
        output_url = upload(output)

        return  {"output": output_url,"status":True} 

      

      #Khushi Sagar- DOWNSAMPLE-Lower Resolution
      elif opr == 'lower_reso':
        output = lower_reso(img)
        print("lower_reso called", output)
        output_url = upload(output)

        return  {"output": output_url,"status":True} 

      #FORAM- Rotation
      elif opr == 'rotate':
        output = combine(img)
        print("rotate called", output)
        output_url = upload(output)

        return  {"output": output_url,"status":True} 


      #Aarti -Laplacian sharpen filter
      elif opr == 'laplacian':
        output = laplacian(img)
        print("laplacian called", output)
        output_url = upload(output)

        return  {"output": output_url,"status":True}



      #KASHISH- RGB to GRAY
      elif opr == 'RgbGray':
        output = RgbGray(img)
        print("RgbGray called", output)
        output_url = upload(output)

        return  {"output": output_url,"status":True}


      
      #Bansi - (Hough transform)
      elif opr == 'LineDetection':
        output = LineDetection(img)
        print("LineDetection called", output)
        output_url = upload(output)

        return  {"output": output_url,"status":True}



      #Keya (Minimum Filter)
      elif opr == 'min_filter':
        output = min_filter(img)
        print("min_filter called", output)
        output_url = upload(output)

        return  {"output": output_url,"status":True}

   #Jinal (Convolution function)
      elif opr == 'gaussianBlur':
        output = gaussianBlur(img)
        print("gaussianBlur called", output)
        output_url = upload(output)
        
        return  {"output": output_url,"status":True}     



     #Zeel Rathi Watershed Function 
      elif opr == 'watershed':
        output = watershed(img)
        print("watershed called", output)
        output_url = upload(output)

        return  {"output": output_url,"status":True}


    


    if (request.method == "GET"):
      print("got ouput")
      try:
        print("get output", output)
        output_url = upload(output)

        return  {"output": output_url}

      except:
        print("No Input Yet")
      

    return "Image Filter"


app.run()
```
